```python
# flawless_ai.py
# RAM optimized, explicitly handles tokens, zero dependency issues

import torch
from llama_index.core import (
    VectorStoreIndex,
    StorageContext,
    ServiceContext,
)


from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import chromadb

# Step 1: Clear and efficient embeddings
embed_model = HuggingFaceEmbedding(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Step 2: Load and index document (minimal data to avoid RAM overflow)
#documents = SimpleDirectoryReader("C:/flawless_ai/documents/").load_data()
documents = SimpleDirectoryReader("C:\\flawless_ai\\documents\\").load_data()



# Create persistent local vector database (ChromaDB)
chroma_client = chromadb.PersistentClient(path="C:/flawless_ai/chroma_db")
vector_store = StorageContext.from_defaults(
    vector_store=chroma_client.get_or_create_collection("flawless_collection")
)

# Setup optimized indexing service context
service_context = ServiceContext.from_defaults(embed_model=embed_model, chunk_size=256)

# Index and store embeddings
index = VectorStoreIndex.from_documents(
    documents, storage_context=vector_store, service_context=service_context
)

# Step 3: Load quantized LLM model to avoid RAM overload
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
)

tokenizer = AutoTokenizer.from_pretrained(
    "mistralai/Mistral-7B-Instruct-v0.2", token=True
)

model = AutoModelForCausalLM.from_pretrained(
    "mistralai/Mistral-7B-Instruct-v0.2",
    quantization_config=quantization_config,
    device_map="auto",
    token=True,
)

llm = HuggingFaceLLM(
    model=model, tokenizer=tokenizer, max_new_tokens=150, context_window=2048
)

# Step 4: Clearly defined retrieval-augmented generation (RAG) query engine
query_engine = index.as_query_engine(llm=llm)

# User query (short to avoid token overflow)
query = "What's the latest Alzheimer's drug approved in 2025?"

# Perform query and clearly print response
response = query_engine.query(query)
print("\nFlawless AI Agent Response:\n", response)
```
